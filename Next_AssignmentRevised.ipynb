```python
import numpy as np
from PIL import Image
import requests
```

```python
from io import BytesIO
import matplotlib.pyplot as plt

def red_filter(image):
    red_filtered_image = np.copy(image)
    red_filtered_image[:, :, 1:] = 0 # Set green and blue channels to zero
    return red_filtered_image

def green_filter(image):
    green_filtered_image = np.copy(image)
    green_filtered_image[:, :, [0, 2]] = 0 # Set red and blue channels to zero
    return green_filtered_image

def blue_filter(image):
    blue_filtered_image = np.copy(image) # Set red and green channels to zero
    blue_filtered_image[:, :, :2] = 0
    return blue_filtered_image

def RG_no_B(image):
    RG_no_B_image = np.copy(image) # Set blue channel to zero
    RG_no_B_image[:, :, 2] = 0
    return RG_no_B_image

def RB_no_G(image):
    RB_no_G_image = np.copy(image) # Set green channel to zero
    RB_no_G_image[:, :, 1] = 0
    return RB_no_G_image

def GB_no_R(image):
    GB_no_R_image = np.copy(image) # Set red channel to zero
    GB_no_R_image[:, :, 0] = 0
    return GB_no_R_image

def display_images(url):
  # Loads images inputted through the URL
    url = input("Please enter the URL of a .jpg file: ")
    response = requests.get(url)
    original_image = Image.open(BytesIO(response.content))
    # Converts image to numpy array
    original_array = np.array(original_image)

  # Applies filters to the arrays
    red_filtered = red_filter(original_array)
    green_filtered = green_filter(original_array)
    blue_filtered = blue_filter(original_array)
    RG_no_B_filtered = RG_no_B(original_array)
    RB_no_G_filtered = RB_no_G(original_array)
    GB_no_R_filtered = GB_no_R(original_array)

  # Combines RGB filtered images
    combined_rgb = np.stack([red_filtered[:,:,0], green_filtered[:,:,1], blue_filtered[:,:,2]], axis=-1)

  # Displays images in an array format
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    axes[0, 0].imshow(original_image)
    axes[0, 0].set_title('Original Image')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(Image.fromarray(red_filtered))
    axes[0, 1].set_title('Red Filtered Image')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(Image.fromarray(green_filtered))
    axes[0, 2].set_title('Green Filtered Image')
    axes[0, 2].axis('off')

    axes[0, 3].imshow(Image.fromarray(blue_filtered))
    axes[0, 3].set_title('Blue Filtered Image')
    axes[0, 3].axis('off')

    axes[1, 0].imshow(Image.fromarray(RG_no_B_filtered))
    axes[1, 0].set_title('R and G, No B')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(Image.fromarray(RB_no_G_filtered))
    axes[1, 1].set_title('R and B, No G')
    axes[1, 1].axis('off')

    axes[1, 2].imshow(Image.fromarray(GB_no_R_filtered))
    axes[1, 2].set_title('G and B, No R')
    axes[1, 2].axis('off')

    axes[1, 3].imshow(Image.fromarray(combined_rgb))
    axes[1, 3].set_title('Combined RGB')
    axes[1, 3].axis('off')

    plt.show()
  # Checks if the original image is the same as the combined RGB image
    original_identical = np.array_equal(original_array, combined_rgb)
    print("Original image is identical to the combined RGB image")
  # Calls the display of the url image
display_images(url)
```
